```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load results CSV
df = pd.read_csv("../../tables/active_learning_runs_v1.csv") 
```

```python
# Plot 1: Accuracy vs. Labeled Samples
plt.figure()
sns.lineplot(data=df, x="total_labeled_samples", y="accuracy_final", hue="strategy_type", marker="o")
plt.title("Accuracy vs. Number of Labeled Samples")
plt.xlabel("Labeled Samples")
plt.ylabel("Final Accuracy")
plt.grid(True)
plt.legend(title="Strategy")
plt.tight_layout()
plt.savefig("../../figures/al_strategy_visualizations/accuracy_vs_samples.png")
```

```python
# Plot 2: Accuracy Std Dev per Strategy
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="accuracy_std")
plt.title("Accuracy Standard Deviation per Strategy")
plt.xlabel("Strategy")
plt.ylabel("Std Deviation")
plt.tight_layout()
plt.savefig("../../figures/al_strategy_visualizations/accuracy_std_per_strategy.png")
```

```python
# Plot 3: Time per Iteration
df["time_per_iter"] = df["duration_sec"] / df["iterations"]
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="time_per_iter")
plt.title("Time per Iteration by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Time per Iteration (s)")
plt.tight_layout()
plt.savefig("../../figures/al_strategy_visualizations/time_per_iteration.png")
```

```python
# Plot 4: Final Accuracy Comparison
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="accuracy_final")
plt.title("Final Accuracy by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Final Accuracy")
plt.tight_layout()
plt.savefig("../../figures/al_strategy_visualizations/final_accuracy_comparison.png")
```
