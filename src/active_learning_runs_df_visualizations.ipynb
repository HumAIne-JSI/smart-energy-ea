```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load results CSV
df = pd.read_csv("../tables/active_learning_runs_v1.csv", sep=";") 
```

```python
df_train_test = df[df["Data Splitting Strategy (extra column)"] == "Train/Test Split"]
df_manual_split = df[df["Data Splitting Strategy (extra column)"] == "Manual Temporal Split"]

# Train/Test Split
plt.figure()
sns.lineplot(data=df_train_test, x="total_labeled_samples", y="accuracy_final", hue="strategy_type", marker="o")
plt.title("Accuracy vs Labeled Samples (Train/Test Split)")
plt.xlabel("Labeled Samples")
plt.ylabel("Final Accuracy")
plt.grid(True)
plt.legend(title="Strategy")
plt.tight_layout()
plt.savefig("../figures/accuracy_vs_samples_train_test_split.png")

# Manual Temporal Split
plt.figure()
sns.lineplot(data=df_manual_split, x="total_labeled_samples", y="accuracy_final", hue="strategy_type", marker="o")
plt.title("Accuracy vs Labeled Samples (Manual Temporal Split)")
plt.xlabel("Labeled Samples")
plt.ylabel("Final Accuracy")
plt.grid(True)
plt.legend(title="Strategy")
plt.tight_layout()
plt.savefig("../figures/accuracy_vs_samples_manual_split.png")
```

```python
# Plot 2: Accuracy Std Dev per Strategy
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="accuracy_std")
plt.title("Accuracy Standard Deviation per Strategy")
plt.xlabel("Strategy")
plt.ylabel("Std Deviation")
plt.tight_layout()
plt.savefig("../figures/accuracy_std_per_strategy.png")
```

```python
# Plot 3: Time per Iteration
df["time_per_iter"] = df["duration_sec"] / df["iterations"]
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="time_per_iter")
plt.title("Time per Iteration by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Time per Iteration (s)")
plt.tight_layout()
plt.savefig("../figures/time_per_iteration.png")
```

```python
# Plot 4: Final Accuracy Comparison
plt.figure()
sns.boxplot(data=df, x="strategy_type", y="accuracy_final")
plt.title("Final Accuracy by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Final Accuracy")
plt.tight_layout()
plt.savefig("../figures/final_accuracy_comparison.png")
```
